```python
import pandas as pd
import numpy as np
import json
import re
import glob
import os

sheet_id = "1gvVTBlKcWl5n_2J0kpxXKNnJMkiwQHkbNqbNy-HD3GA"

def read(sheet_name):
    sheet_url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}'
    return pd.read_csv(sheet_url,encoding='utf-8')
```

```python
project=[]
dfs=[]
for k in range(1,11):
    df=read(f'a{k}')
    dfs.append(df)
  
    s={}
    for i, r in df.iterrows():
        if not 'description' in r:
            print('***************************** no description', i, f'a{k}', r)
            break
        if not pd.isna(r['description']):
            s[r['field']]=r['description']
            print(r['field'], r['description'])
    project.append(s)
```

```python
os.chdir('images.ai')
files=glob.glob('*')
os.chdir('..')
print(files)
```

```python
for i in range(len(project)):
    p=project[i]
    print(f'# {i+1}')
    for e in p:
        print(f'## {e}')
        lines=p[e].split('\n')
        no=1
        for line in lines:
            m=re.search('^(\- *)(.*)', line.strip())
            if m: 
                print(f'{no}. {m.group(2)}')
                no+=1
            else:
                m=re.search('^(\. *)(.*)', line.strip())
                if m: 
                    print(f'    + {m.group(2)}')
                else: print(p[e])
        if e=='산출물':
            print('## 산출물 샘플')
            for image in files:
                if re.search(f'^{i+1}-', image):
                    print(f'![{image}](https://github.com/ekyuho/magicecole/blob/main/images.ai/{image}?raw=true)')
```
