```python
projects={}
for i, r in df.iterrows():
    index=r['index']
    key=r['key']
    value=r['value']
    
    if index not in projects: 
        print('found new project', index)
        projects[index]={}
    
    projects[index][key]=value
```

```python
os.chdir('images')
files=glob.glob('*')
os.chdir('..')
print(files)
```

```python
target=['a1','a2','a3','a4','a5','a6','a7','a8','a9','a10']
single=['구분','제목','시간']

for index in target:
    print(f'# 프로젝트코드: {index}')
    p=projects[index]
    print('|항목|내용|')
    print('|--|--|')
    for x in single:
        print(f'|{x}|{p[x]}|')
    
    for key in p:
        if key in single: continue
        print(f'## {key}:')
        if not pd.isna(p[key]):
            lines=p[key].split('\n')
            no=1
            for line in lines:
                m=re.search('^(\- *)(.*)', line.strip())
                if m: 
                    print(f'{no}. {m.group(2)}')
                    no+=1
                else:
                    m=re.search('^(\. *)(.*)', line.strip())
                    if m: 
                        print(f'    + {m.group(2)}')
                    else: print(p[key])
                    
        if key=='결과물 예시':
            print('## 결과물 예시')
            for image in files:
                if re.search(f'^{index}-', image):
                    print(f'![{image}](https://github.com/ekyuho/magicecole/blob/main/images/{image}?raw=true)')
```
