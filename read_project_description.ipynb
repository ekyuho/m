```python
import pandas as pd
import numpy as np
import json
import re

sheet_id = "sheet_id"

def read(i):
    sheet_name = f"{i}"
    sheet_url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}'
    return pd.read_csv(sheet_url, encoding='utf-8')
```

```python
project=[]
for k in range(1,10):
    df=read(k)
    s={}
    for i, r in df.iterrows():
        if not pd.isna(r['description']):
            s[r['field']]=r['description']
    project.append(s)
```

```python
for i in range(len(project)):
    p=project[i]
    print(f'# {i+1}')
    for e in p:
        print(f'## {e}')
        lines=p[e].split('\n')
        no=1
        for line in lines:
            m=re.search('^(\- *)(.*)', line.strip())
            if m: 
                print(f'{no}. {m.group(2)}')
                no+=1
            else:
                m=re.search('^(\. *)(.*)', line.strip())
                if m: 
                    print(f'    + {m.group(2)}')
                else: print(p[e])
```
